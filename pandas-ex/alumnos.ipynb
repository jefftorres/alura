# 2. Tratamiento

```python
alumnos.isnull().sum()
```

```python
df = alumnos.fillna(0)
df
```

```python
est = ["Alicia", "Carlos"]

df.drop(df.query('@est in Nombre').index, inplace=True)
df
```

```python
filter1 = df['Aprobado'] == 'True'

aprobados = df[filter1]
aprobados
```

```python
aprobados.to_csv('alumnos_aprobados.csv', index=False)
```

```python
aprobados_rev = aprobados.where(aprobados['Nota'] != 7., 8.)
aprobados_rev
```

```python
df['Puntos_extras'] = df['Nota'] * .4
df['Notas_finales'] = df.Nota + df.Puntos_extras
df.head()
```

```python
df['Aprobado_final'] = df['Notas_finales'].apply(lambda x: 'True' if x >= 7. else 'False')
df.head()
```

```python
select = (df['Aprobado'] == 'False') & (df['Aprobado_final'] == 'True')
df[select]
```
