```python
import numpy as np
import pandas as pd
```

```python
ventas = pd.read_json('data/dados_vendas_clientes.json')

ventas = pd.json_normalize(ventas['dados_vendas'])
ventas.head()
```

```python
ventas = ventas.explode(ventas.columns.tolist()[1:])
ventas = ventas.reset_index(drop=True)
ventas.head()
```

```python
ventas.dtypes
```

```python
ventas['Valor da compra'] = ventas['Valor da compra'].apply(lambda p: p.replace(',', '.').replace('R$ ', '').strip())
ventas['Valor da compra'] = ventas['Valor da compra'].astype(np.float64)
ventas.dtypes
```

```python
ventas.head()
```

```python
ventas['Cliente'] = ventas['Cliente'].str.lower()
ventas.head()
```

```python
ventas['Cliente'] = ventas['Cliente'].str.replace('[^a-z ]', '', regex=True).str.strip()
ventas.head()
```

```python
ventas['Data de venda'] = pd.to_datetime(ventas['Data de venda'], format='%d/%m/%Y')
ventas.head()
```

El objetivo de este proyecto es analizar los resultados de un evento con los clientes de una empresa de venta online. Se recopiló un conjunto de datos que contiene los clientes que gastaron más en productos dentro de los 5 días posteriores a la venta, que es la duración del evento. Este análisis identificará al cliente con la mayor compra esa semana, quien recibirá un premio de la tienda, y posteriormente, puede ayudar a la empresa a crear nuevas estrategias para atraer más clientes.
